```python
#normalization function
def  featureNormalize(X):
    
    X_norm = X.copy()
    mu = np.zeros(X.shape[1])
    sigma = np.zeros(X.shape[1])

    mu=np.mean(X,axis=0)
    sigma=np.std(X,axis=0)
    
    X_norm= (X_norm-mu)/sigma

    return X_norm, mu, sigma
```

```python
#normalizing the train , cv and test sets and adding a column = 1's for theta 0
X_train_norm, mu, sigma = featureNormalize(X_train)
X_cv_norm = (X_cv - mu) / sigma
X_test_norm = (X_test - mu) / sigma

m1 = y_train.size
m2 = y_cv.size
m3 = y_test.size
X_train_norm = np.concatenate([np.ones((m1, 1)), X_train_norm], axis=1)
X_cv_norm = np.concatenate([np.ones((m2, 1)), X_cv_norm], axis=1)
X_test_norm = np.concatenate([np.ones((m3, 1)), X_test_norm], axis=1)
```

# Linear Regression

```python
#Cost function for multivariables
def computeCostMulti(X, y, theta):

    m = y.shape[0] # number of training examples
    J = 0
        
    J=(1/(2*m))*np.dot(((np.dot(X,theta)-y).T),np.dot(X,theta)-y)
 
    return J
```

```python
#gradient descent functioon for multivariables
def gradientDescentMulti(X, y, theta, alpha, num_iters):

    # Initialize some useful values
    m = y.shape[0] # number of training examples
    
    # make a copy of theta, which will be updated by gradient descent
    theta = theta.copy()
    
    J_history = []
    
    for i in range(num_iters):
        sumofh0x=np.dot(X,theta)
        theta=theta-((alpha/m)*(np.dot(X.T,sumofh0x-y)))       
        
        # save the cost J in every iteration 
        J_history.append(computeCostMulti(X, y, theta))
    
    return theta, J_history
```

```python
#normal equation function
def normalEqn(X, y):

    theta_norm = np.zeros(X.shape[1])
    theta_norm = np.dot((np.linalg.inv(np.dot(X.T,X))),np.dot(X.T,y))

    return theta_norm
```

```python
#applying the cost function and gradient descent function on the linear hypothesis
alpha = 0.1
num_iters = 400

# init theta and run gradient descent
theta1 = np.zeros(cor.size)
theta1, J_history = gradientDescentMulti(X_train_norm, y_train, theta1, alpha, num_iters)

# Plot the convergence graph
pyplot.plot(np.arange(len(J_history)), J_history, lw=2)
pyplot.xlabel('Number of iterations')
pyplot.ylabel('Cost J')

# Display the gradient descent's result
print('theta computed from gradient descent: {:s}'.format(str(theta1)))

J_train = computeCostMulti(X_train_norm, y_train, theta1) 
J_cv = computeCostMulti(X_cv_norm, y_cv, theta1)
J_test = computeCostMulti(X_test_norm, y_test, theta1)    

J_train,J_cv, J_test
```

```python
# Calculate the parameters from the normal equation
theta_norm = normalEqn(X_train_norm, y_train);

# Display normal equation's result
print('Theta computed from the normal equations: {:s}'.format(str(theta_norm)));

J_train_norm = computeCostMulti(X_train_norm, y_train, theta_norm) 
J_cv_norm = computeCostMulti(X_cv_norm, y_cv, theta_norm)
J_test_norm = computeCostMulti(X_test_norm, y_test, theta_norm)    

J_train_norm,J_cv_norm, J_test_norm
```

# Regularization

```python
#cost function for regularization
def CostMultiReg(X, y, theta, lambda_):

    m = y.shape[0] # number of training examples
    J = 0
    sumofh0x=np.dot(X,theta)
    theta_reg = theta.copy()
    theta_reg [0] = 0
    
    J=(1 / (2 * m)) * np.sum(np.square(sumofh0x - y)) + (lambda_ / (2 * m)) * np.sum(np.square(theta_reg))
 
    return J
```

```python
#gradient descent and cost function for regularization
def RegCostFunction(X, y, theta, alpha, num_iters, lambda_):
    # Initialize some useful values
    m = y.shape[0] # number of training examples
    
    # make a copy of theta, which will be updated by gradient descent
    grad = np.zeros(theta.shape)
    theta = theta.copy()
    theta_reg = theta.copy()
    theta_reg [0] = 0
    J_history = []
    
    for i in range(num_iters):
        sumofh0x=np.dot(X,theta)
        theta=theta-((alpha/m)*((np.dot(X.T,sumofh0x-y))+(lambda_*theta_reg)))       
        
        # save the cost J in every iteration 
        J_history.append((1 / (2 * m)) * np.sum(np.square(sumofh0x - y)) + (lambda_ / (2 * m)) * np.sum(np.square(theta_reg)))
    
    return theta, J_history
```

```python
#K-Fold method with gradient descent 
i = 1
alpha = 0.1
lambda_= 0.5
num_iters = 400
theta_Fold_Reg = np.zeros(cor.size)
kfold = KFold(10)
K= 10
X = new_data.drop(["price"],axis=1)
Y = new_data["price"]
J_train_Fold = 0
J_test_Fold = 0
J_test_total = np.zeros(K)

for train, test in kfold.split(X):
    
    X_train_Fold = X.iloc[train]
    Y_train_Fold = Y.iloc[train]
    Y_train_Fold_size = Y_train_Fold.size
    
    X_test_Fold = X.iloc[test]
    Y_test_Fold = Y.iloc[test]
    Y_test_Fold_size = Y_test_Fold.size

    X_train_Fold_norm, mu_Fold, sigma_Fold = featureNormalize(X_train_Fold)
    X_test_Fold_norm = (X_test_Fold - mu_Fold)/(sigma_Fold)    
    X_train_Fold_norm = np.concatenate([np.ones((Y_train_Fold_size,1)), X_train_Fold_norm], axis = 1 )
    X_test_Fold_norm =  np.concatenate([np.ones((Y_test_Fold_size,1)), X_test_Fold_norm],  axis = 1 )
    
    theta_Fold_Reg, J_history = RegCostFunction(X_train_Fold_norm, Y_train_Fold, theta_Fold_Reg, alpha, num_iters,  lambda_)
    
    #Printing the tag for each fold
    print('Fold No. = ',i)
    
    J_train_Fold_Reg = CostMultiReg(X_train_Fold_norm, Y_train_Fold, theta_Fold_Reg, lambda_) 
    J_test_Fold_Reg = CostMultiReg(X_test_Fold_norm, Y_test_Fold, theta_Fold_Reg, lambda_)    
    J_test_total[i-1] = J_test_Fold_Reg

    print('J_train = ',J_train_Fold_Reg)
    print('J_test = ',J_test_Fold_Reg)
    
    #Incrementing the folds counter
    i = i + 1
J_test_average = np.sum(J_test_total)/K   
print('J Average', J_test_average)
```
